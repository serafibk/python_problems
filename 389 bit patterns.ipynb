```python
import numpy as np
import time


#generating binary strings
def decToBin(num):
    return bin(num).replace('0b','')

#summing a bit string  (number of ones)
def sumBitString(N,n):
    N=list(N)
    
    sum=0
    
    for i in range(n):
        sum = sum+int(N[i])
        
    return sum

#determine if N2 can follow N1 in valid traingle array
def isValidStep(N1,N2,n):
    sum1=0
    sum2=0
    
    N1 = list(N1)
    N2 = list(N2)
    
    for i in range(n):
        sum1 = sum1+int(N1[i])
        sum2 = sum2+int(N2[i])
        
        if(sum2>sum1):
            return False
        if(sum1-sum2>=2):
            return False
    if(sum1-sum2 != 1):
        return False
    return True

#create an array full of all possible binary numbers up to 2^n
def generateBinaryArray(n):
    numArr = []
    for i in range(2**n):
        num = decToBin(i)
        
        #fix total length
        while len(num) != n:
            num = '0'+num
        numArr.append(num)
    return numArr

#counts number of arrays that have N as top row
count = 0

#TRY TO WRITE MEMOIZATION TO STORE COUNTS OF SMALLER ARRAYS TO CUT DOWN RUNTIME
def countNumArrays(N):
    binArr = generateBinaryArray(len(N))
    global count
    
    for a in binArr:
        valid=isValidStep(N,a,len(N))
        sumA = sumBitString(a,len(a))
        if(valid):
            if(sumA==1):
                count = count+1
                #dont recurse further
                continue
            countNumArrays(a)        
    return count


#STRICTLY INCREASING ARRAYS 

#starting row
N = '11'
start = time.time()
print('n=2: ', countNumArrays(N))
end = time.time()
print('time elapsed: ', end-start)
count = 0

N='111'
start = time.time()
print('n=3: ', countNumArrays(N))
end = time.time()
print('time elapsed: ', end-start)
count = 0

N='1111'
start = time.time()
print('n=4: ', countNumArrays(N))
end = time.time()
print('time elapsed: ', end-start)
count = 0

N='11111'
start = time.time()
print('n=5: ', countNumArrays(N))
end = time.time()
print('time elapsed: ', end-start)
count = 0

N='111111'
start = time.time()
print('n=6: ', countNumArrays(N))
end = time.time()
print('time elapsed: ', end-start)
count = 0

```
