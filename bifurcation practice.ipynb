```python
import numpy as np
import matplotlib.pyplot as plt

#function of population growth
def nextPop (rate, currentPop):
    return currentPop*rate*(1-currentPop)

#define rate and initial pop values and initialize array to store next pops for each initail pop
rate = np.arange(2.5,4,0.003)
initialPop = 0.78
popArr = np.arange(0,50,.1)
eps = 0.015

#create a figure to plot convergences
figure = plt.subplots(1, figsize=(11,5))

#fill nextPop with values from funciton
num = 0;
for r in rate:
    popArr[0] = initialPop
    for p in range(0,499):
        popArr[p+1] = nextPop(r,popArr[p])     
    p = 1
    cInd = 1
    convergePop = [popArr[499]]
    size = 1
    while(abs(popArr[499]- popArr[499-p])>= eps and p <500):
        convergePop.append(popArr[499-p])
        cInd = cInd + 1 
        p=p+1
        size = size+1    
    tempRate = np.ones(size)*r
    plt.scatter(tempRate,convergePop,marker='h')
    plt.title('bifurcation graph of population growth')
    plt.xlabel('growth rate r')
    plt.ylabel('converging population x')
```
